```python
import pandas as pd

gen = pd.read_csv('music_raw_data.csv')
```

```python
gen.columns=['age','gender', 'genre']
gen.head()
```

```python
for i, x in gen.iterrows():
    if gen.loc[i, 'gender'] <= 15: # high
        gen.loc[i, 'gender'] =  gen.loc[i, 'gender'] + 15

gen.head()
```

```python
for i, x in gen.iterrows():
    if gen.loc[i, 'gender'] <= 35:
        gen.loc[i, 'gender'] = 1 #female
    else: # low
        gen.loc[i, 'gender'] = 0 #male

gen.head()
```

```python
for i, x in gen.iterrows():
    if gen.loc[i, 'gender'] == 1: # high
        gen.loc[i, 'gender'] = 'female' #female
    elif gen.loc[i, 'gender'] == 0: # low
        gen.loc[i, 'gender'] = 'male' #male

gen.head()
```

```python
for i, x in gen.iterrows():
    if gen.loc[i, 'age'] <= 15: # high
        gen.loc[i, 'age'] =  gen.loc[i, 'age'] + 15

gen.head()
```

```python
for i, x in gen.iterrows():
    if gen.loc[i, 'age'] >= 0  and gen.loc[i, 'age'] <= 20 : 
        gen.loc[i, 'genre'] = 'rock' 
    elif gen.loc[i, 'age'] >= 21  and gen.loc[i, 'age'] <= 40 and gen.loc[i, 'gender'] == 'female':
        gen.loc[i, 'genre'] = 'pop'   
    elif gen.loc[i, 'age'] >= 21  and gen.loc[i, 'age'] <= 40 and gen.loc[i, 'gender'] == 'male':
        gen.loc[i, 'genre'] = 'jazz' 
    elif gen.loc[i, 'age'] > 40 and gen.loc[i, 'gender'] == 'female':
        gen.loc[i, 'genre'] = 'ambient' 
    elif gen.loc[i, 'age'] > 40 and gen.loc[i, 'gender'] == 'male':
        gen.loc[i, 'genre'] = 'classical'

gen
```

```python
gen.to_csv('music_data_preproc.csv', header=False, index=False) 
```
